```python
import os
import pickle
import numpy as np
import seaborn as sns
from tqdm import tqdm
import tensorflow.keras as K
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from sklearn.model_selection import StratifiedShuffleSplit
from tensorflow.keras.preprocessing.image import img_to_array

parent_dir = '/content/drive/MyDrive/Skin_disease/dataset_dir/train_set'

# Print the list of subdirectories
subdirs = [d for d in os.listdir(parent_dir) if os.path.isdir(os.path.join(parent_dir, d))]
print("Subdirectories:", subdirs)

features = []
labels = []

for subdir in subdirs:
    subdir_path = os.path.join(parent_dir, subdir)
    img_paths = []

    for root, dirs, files in os.walk(subdir_path):
        for file in files:
            img_paths.append(os.path.join(root, file))

    for i in tqdm(img_paths, desc=f"Processing {subdir}"):
        img = image.load_img(i, target_size=(299, 299))
        x = img_to_array(img)
        x = K.applications.xception.preprocess_input(x)
        features.append(x)
        labels.append(subdir)

features = np.array(features)
labels = np.array(labels)

# Shuffle the data
indices = np.random.permutation(len(features))
features = features[indices]
labels = labels[indices]

# Split the data
x_train, y_train, rx_test, ry_test, x_val, y_val, x_test, y_test = [], [], [], [], [], [], [], []

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=0)
for train_index, r_index in split.split(features, labels):
    x_train, rx_test = features[train_index], features[r_index]
    y_train, ry_test = labels[train_index], labels[r_index]

split = StratifiedShuffleSplit(n_splits=1, test_size=0.15, random_state=0)
for val_index, test_index in split.split(rx_test, ry_test):
    x_val, x_test = rx_test[val_index], rx_test[test_index]
    y_val, y_test = ry_test[val_index], ry_test[test_index]

# Save the processed data into pickle files
f = open("/content/drive/MyDrive/Skin_disease/x_train", "wb")
pickle.dump(x_train, f)
f.close()

f = open("/content/drive/MyDrive/Skin_disease/y_train", "wb")
pickle.dump(y_train, f)
f.close()

f = open("/content/drive/MyDrive/Skin_disease/x_val", "wb")
pickle.dump(x_val, f)
f.close()

f = open("/content/drive/MyDrive/Skin_disease/y_val", "wb")
pickle.dump(y_val, f)
f.close()

f = open("/content/drive/MyDrive/Skin_disease/x_test", "wb")
pickle.dump(x_test, f)
f.close()

f = open("/content/drive/MyDrive/Skin_disease/y_test", "wb")
pickle.dump(y_test, f)
f.close()
```

```python
import pickle
import numpy as np
from tqdm import tqdm
from skimage.transform import rotate

x_train = pickle.load(open('train_val_test_sets6/after_aug/x_train', 'rb'))
y_train = pickle.load(open('train_val_test_sets6/after_aug/y_train', 'rb'))
x_val = pickle.load(open('/content/drive/MyDrive/Skin_disease/x_val', 'rb'))
y_val = pickle.load(open('/content/drive/MyDrive/Skin_disease/y_val', 'rb'))
x_test = pickle.load(open('/content/drive/MyDrive/Skin_disease/x_test', 'rb'))
y_test = pickle.load(open('/content/drive/MyDrive/Skin_disease/y_test', 'rb'))

x_t = len(x_train)
x_copy = x_train.copy()
count = 0
ecz, ker, ros = 0, 0, 0
for i in tqdm(range(x_t)):
    if y_train[i] == 0:
        x_train = np.append(x_train, [rotate(x_copy[i], angle=45, mode='wrap')], axis=0)
        y_train = np.append(y_train, 0)

    elif y_train[i] == 2:
        if ecz % 3 == 0:
            x_train = np.append(x_train, [rotate(x_copy[i], angle=45, mode='wrap')], axis=0)
            x_train = np.append(x_train, [np.flipud(x_copy[i])], axis=0)
            for j in range(2):
                y_train = np.append(y_train, 2)
        ecz += 1

    elif y_train[i] == 3:
        if ker % 6 == 0:
            x_train = np.append(x_train, [rotate(x_copy[i], angle=45, mode='wrap')], axis=0)
            x_train = np.append(x_train, [np.flipud(x_copy[i])], axis=0)
            for j in range(2):
                y_train = np.append(y_train, 3)
        ker += 1

    elif y_train[i] == 4:
        x_train = np.append(x_train, [rotate(x_copy[i], angle=45, mode='wrap')], axis=0)
        x_train = np.append(x_train, [np.flipud(x_copy[i])], axis=0)
        x_train = np.append(x_train, [np.fliplr(x_copy[i])], axis=0)
        x_train = np.append(x_train, [rotate(np.fliplr(x_copy[i]), angle=270, mode='wrap')], axis=0)
        for j in range(4):
            y_train = np.append(y_train, 4)

    elif y_train[i] == 5:
        if ros % 9 == 0:
            x_train = np.append(x_train, [rotate(x_copy[i], angle=45, mode='wrap')], axis=0)
            x_train = np.append(x_train, [np.flipud(x_copy[i])], axis=0)
            for j in range(2):
                y_train = np.append(y_train, 5)
        ros += 1

indices = np.random.permutation(len(x_train))
x_train = x_train[indices]
y_train = y_train[indices]

f = open("/content/drive/MyDrive/Skin_disease/x_train", "wb")
pickle.dump(x_train, f)
f.close()
f = open("/content/drive/MyDrive/Skin_disease/y_train", "wb")
pickle.dump(y_train, f)
f.close()

f = open("/content/drive/MyDrive/Skin_disease/x_val", "wb")
pickle.dump(x_val, f)
f.close()
f = open("/content/drive/MyDrive/Skin_disease/y_val", "wb")
pickle.dump(y_val, f)
f.close()

f = open("/content/drive/MyDrive/Skin_disease/x_test", "wb")
pickle.dump(x_test, f)
f.close()
f = open("/content/drive/MyDrive/Skin_disease/y_test", "wb")
pickle.dump(y_test, f)
f.close()
```

```python
import pickle
from sklearn.preprocessing import LabelEncoder
from keras import Model
import tensorflow.keras as K
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint
from keras.layers import GlobalAveragePooling2D, BatchNormalization, Dense, Dropout

# Load your data
x_train = pickle.load(open('/content/drive/MyDrive/Skin_disease/x_train', 'rb'))
y_train = pickle.load(open('/content/drive/MyDrive/Skin_disease/y_train', 'rb'))
x_val = pickle.load(open('/content/drive/MyDrive/Skin_disease/x_val', 'rb'))
y_val = pickle.load(open('/content/drive/MyDrive/Skin_disease/y_val', 'rb'))
x_test = pickle.load(open('/content/drive/MyDrive/Skin_disease/x_test', 'rb'))
y_test = pickle.load(open('/content/drive/MyDrive/Skin_disease/y_test', 'rb'))

# Convert string labels to integers
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)
# Repeat this for y_test if needed

num_classes = len(label_encoder.classes_)

base_model = K.applications.Xception(include_top=False,
                                     weights='imagenet',
                                     input_tensor=None,
                                     input_shape=(299, 299, 3),
                                     pooling=None,
                                     classifier_activation="softmax")

base_model.trainable = False

inputs = K.Input(shape=(299, 299, 3))

x = base_model(inputs, training=False)
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

# Use softmax activation and adjust the number of output neurons
outputs = Dense(num_classes, activation='softmax')(x)

model = Model(inputs, outputs)

opt = Adam(learning_rate=0.001)
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])

# Checkpoints
acc_checkpoint = ModelCheckpoint("/content/drive/MyDrive/Skin_disease/second_train/max_acc", monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
loss_checkpoint = ModelCheckpoint("/content/drive/MyDrive/Skin_disease/second_train/min_loss", monitor='val_loss', verbose=1, save_best_only=True, mode='min')
callbacks_list = [acc_checkpoint, loss_checkpoint]

# Training
hist = model.fit(x_train, y_train_encoded, epochs=10, validation_data=(x_val, y_val_encoded), batch_size=32, callbacks=callbacks_list)

# Save training history
with open('/content/drive/MyDrive/Skin_disease/second/hist', 'wb') as file_pi:
    pickle.dump(hist.history, file_pi)
```

```python
import numpy as np
import tensorflow.keras as K
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt
from IPython.display import display
import ipywidgets as widgets
import io
import os

model_path = "/content/drive/MyDrive/Skin_disease/second_train/max_acc"
types = ['BA- cellulitis', 'BA-impetigo', 'FU-athlete-foot', 'FU-nail-fungus',
         'FU-ringworm', 'PA-cutaneous-larva-migrans', 'VI-chickenpox', 'VI-shingles']

try:
    model = K.models.load_model(model_path)
except Exception as e:
    print(f"Error loading the model: {e}")
    exit()

upload = widgets.FileUpload(accept='image/*', multiple=False)

def predict_skin_disease(change):
    uploaded_filename = next(iter(upload.value))
    content = upload.value[uploaded_filename]['content']
    img = image.load_img(io.BytesIO(content), target_size=(299, 299))
    x = img_to_array(img)
    x = K.applications.xception.preprocess_input(x)

    prediction = model.predict(np.expand_dims(x, axis=0))[0]

    print("Predicted probabilities:", prediction)

    result = [(types[i], float(prediction[i]) * 100.0) for i in range(len(prediction))]
    result.sort(reverse=True, key=lambda x: x[1])

    for j in range(min(6, len(result))):
        (class_name, prob) = result[j]
        print(f"Top {j + 1} ====================")
        print(f"{class_name}: {prob:.2f}%")

    plt.imshow(img)
    plt.axis('off')
    plt.show()

upload.observe(predict_skin_disease, names='value')
display(upload)
```

```python
import os
import numpy as np
from tqdm import tqdm
import tensorflow.keras as K
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt

from google.colab import files

model_path = "/content/drive/MyDrive/Skin_disease/second_train/max_acc.h5"
types = ['BA- cellulitis', 'BA-impetigo', 'FU-athlete-foot', 'FU-nail-fungus',
         'FU-ringworm', 'PA-cutaneous-larva-migrans', 'VI-chickenpox', 'VI-shingles']

try:
    model = K.models.load_model(model_path)
except Exception as e:
    print(f"Error loading the model: {e}")
    exit()

def predict_image(fname):
    img = image.load_img(fname, target_size=(299, 299))
    x = img_to_array(img)
    x = K.applications.xception.preprocess_input(x)

    prediction = model.predict(np.expand_dims(x, axis=0))[0]

    print(f'Image name: {fname}')
    print("Predicted probabilities:", prediction)

    max_prob = max(prediction)
    if max_prob < 0.1:
        print("Unable to predict. Confidence level below 50%.")
        return None

    result = [(types[i], float(prediction[i]) * 100.0) for i in range(len(prediction))]
    result.sort(reverse=True, key=lambda x: x[1])

    predictions = []
    for j in range(min(6, len(result))):
        (class_name, prob) = result[j]
        predictions.append({"class_name": class_name, "probability": prob})

    return predictions

uploaded = files.upload()
for fname in uploaded.keys():
    predictions = predict_image(fname)
    if predictions is not None:
        for prediction in predictions:
            print(f"{prediction['class_name']}: {prediction['probability']:.2f}%")
        img = image.load_img(fname)
        plt.imshow(img)
        plt.show()
```
